```python
import pandas as pd
from datetime import datetime
```

```python
covid_data = pd.read_csv("owid-covid-data.csv")
covid_data
```

```python
covid_data.info()
```

```python
#перевела дату у формат дейттайм для зручності у подальшій роботі(наприклад, для завд № 3)
covid_data['date'] = pd.to_datetime(covid_data.date, format='%Y-%m-%d')
```

```python
# перевіримо
covid_data.info()
```

#### `1. Яка загальна кількість хворих зафіксована на Філіппінах на 30.08.2020? `

```python
covid_data_philippines = covid_data[(covid_data["date"] == "2020-08-30") & (covid_data["location"] == "Philippines")]
covid_data_philippines
```

```python
total_cases_philippines = covid_data_philippines["total_cases"].to_list()
total_cases_philippines
```

#### `2. Коли був зафіксований найбільший приріст хворих за тиждень в Україні? `

```python
covid_data_ukraine = covid_data.loc[covid_data.location == 'Ukraine', ['location', 'date', 'total_cases', 'new_cases']]
covid_data_ukraine
```

```python
# для того, щоб дізнатися приріст хворих за тиждень, додаю колонку "номер тижня", за якою потім сгрупую
covid_data_ukraine['week'] = covid_data_ukraine['date'].dt.week
```

```python
covid_data_ukraine.head()
```

```python
# групую за номером тижня і датою(бо дані за два роки), вивожу суму нових випадків за тиждень, сортую від більшого, до меншого
covid_data_ukraine_max = covid_data_ukraine.groupby(['week', 'date']).agg({'new_cases': 'sum'}).sort_values('new_cases', ascending=False)
covid_data_ukraine_max
```

Бачимо, що найбільший приріст хворих був за 5 тиждень 2022 року

```python
# виведемо цей тиждень
covid_data_ukraine['year'] = covid_data_ukraine['date'].dt.strftime('%Y')
```

```python
covid_data_ukraine.loc[(covid_data_ukraine['week'] == 5) & (covid_data_ukraine['year'] == '2022')]
```

#### `3. Порахуйте, чи відповідають дані по загальній кількості випадків за 21.08.2020 по світу сумі випадків по країнах за цю дату?`

```python
covid_data_21_08_2020.total_cases.sum()
```

#### `4. Запишіть в текстовий файл найбільший показник по випадкам по кожній з країн.`
